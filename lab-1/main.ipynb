```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = [10 ** 1, 10 ** 2, 10 ** 3, 10 ** 4]
```

```python
def gen_uniform_sample(n):
    sample = np.random.uniform(0, 1, n)
    return sample
```

```python
def make_cdf(sample):
    sorted_sample = np.sort(sample)
    
    x = np.linspace(0, 1, 1000)
    y = [np.mean(sorted_sample <= r) for r in x]

    return x, y
```

```python
plt.figure(figsize=(12, 8))

for i, n in enumerate(N):
    sample = gen_uniform_sample(n)
    
    x, y = make_cdf(sample)
    
    plt.subplot(2, 2, i+1)
    plt.plot(x, y, label=' ', linewidth=2)
    plt.plot([0, 1], [0, 1], 'r--', label=' ', linewidth=2)
    plt.title(f'n = {n}')
    plt.xlabel('x')
    plt.ylabel('F(x)')
    plt.legend()
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))

for i, n in enumerate(N):
    sample = gen_uniform_sample(n)
    bins = n // 10

    plt.subplot(2, 2, i+1)
    plt.hist(sample, bins=bins, density=True, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title(f'   {len(sample)}')
    plt.xlabel('')
    plt.ylabel(' ')
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
